```python
def get_vocabulary(train_data):
  special_tokens = [
        '[PAD]', '[UNK]',
        '<MeasuredEntity>', '</MeasuredEntity>',
        '<MeasuredProperty>', '</MeasuredProperty>',
        '<Qualifier>', '</Qualifier>',
        '<Quantity>', '</Quantity>'
    ]

    # Get unique lemmas from the training data
  unique_lemmas = train_data['lemma'].astype(str).unique()

    # Create the vocabulary: special tokens followed by unique lemmas
    # Filter out any lemmas that might be accidentally one of the special tokens
  lemma_vocab = [lemma for lemma in unique_lemmas if lemma not in special_tokens]

    # Combine to form the full vocabulary
  vocab = special_tokens + sorted(lemma_vocab)

    # Build the word-to-index dictionary
  word2idx = {word: idx for idx, word in enumerate(vocab)}

  return vocab, word2idx
```

```python
def integrate_sentences(data):
    # Group the DataFrame by document and sentence to process one sentence at a time
    sentences_grouped = data.groupby(['docId', 'sentId'])

    relation_candidates = []

    # Iterate over each sentence
    for (docId, sentId), sentence_data in sentences_grouped:
        # Get all unique entities in the sentence (annotId)
        all_entities = sentence_data['annotId'].dropna().unique()

        # Group tokens by annotId to easily extract entity spans and types
        entity_spans = {}
        for annotId in all_entities:
            entity_tokens = sentence_data[sentence_data['annotId'] == annotId]

            # The entity type is the part of the 'label' after 'B-' or 'I-'
            entity_type = entity_tokens['label'].iloc[0].split('-')[-1]

            # Get the starting and ending indices in the sentence tokens for the entity
            start_index = entity_tokens.index[0]
            end_index = entity_tokens.index[-1]

            # Store the entity details
            entity_spans[annotId] = {
                'type': entity_type,
                'start': start_index,
                'end': end_index,
                'tokens': entity_tokens['lemma'].tolist(),
                # rel contains the annotId of the Quantity it's linked to.
                # Since 'nan' is float('nan'), we convert it to str to use a standard value for comparison.
                'rel_to': sentence_data.loc[start_index, 'rel'] if pd.notna(sentence_data.loc[start_index, 'rel']) else None
            }

        # Separate Quantity entities from other entities
        quantity_entities = {ann_id: details for ann_id, details in entity_spans.items() if details['type'] == 'Quantity'}
        other_entities = {ann_id: details for ann_id, details in entity_spans.items() if details['type'] != 'Quantity'}

        # Form all possible pairs (Quantity, Other)
        for quantityId, quantity_details in quantity_entities.items():
            for otherId, other_details in other_entities.items():

                # Check for HasQuantity relation: the 'rel' of the other entity points to the quantityId
                # Note: The provided example table suggests the 'rel' column on the *other* entity token
                # contains the annotId of the linked *Quantity*.
                is_related = (other_details['rel_to'] == quantityId)

                # --- Construct the tagged sentence ---

                # Get the base list of lemmas for the sentence
                sentence_lemmas = sentence_data['lemma'].tolist()

                # Identify which entity is first in the sentence to tag them correctly in order.
                # Use the global index in the original data to determine order within the sentence.

                # Create a list of (index, tag_type, is_start_tag, entity_type) tuples
                tagging_points = []

                # 1. Tags for Quantity entity
                q_start_idx = quantity_details['start'] - sentence_data.index[0] # Relative index in sentence_lemmas
                q_end_idx = quantity_details['end'] - sentence_data.index[0] # Relative index
                q_tag = f'<{quantity_details["type"]}>'
                q_end_tag = f'</{quantity_details["type"]}>'

                tagging_points.append((q_start_idx, q_tag, True, quantity_details['type']))
                tagging_points.append((q_end_idx + 1, q_end_tag, False, quantity_details['type']))

                # 2. Tags for the Other entity
                o_start_idx = other_details['start'] - sentence_data.index[0]
                o_end_idx = other_details['end'] - sentence_data.index[0]
                o_tag = f'<{other_details["type"]}>'
                o_end_tag = f'</{other_details["type"]}>'

                tagging_points.append((o_start_idx, o_tag, True, other_details['type']))
                tagging_points.1((o_end_idx + 1, o_end_tag, False, other_details['type']))

                # Sort the tagging points: primary key is index (position),
                # secondary key: start tags (True) before end tags (False) at the same position.
                tagging_points.sort(key=lambda x: (x[0], x[2]))

                tagged_lemmas = []
                last_index = 0

                for index, tag, is_start_tag, entity_type in tagging_points:
                    if index > last_index:
                        # Append lemmas between the last tag point and the current tag point
                        tagged_lemmas.extend(sentence_lemmas[last_index:index])

                    # Append the tag
                    tagged_lemmas.append(tag)

                    last_index = index

                # Append any remaining lemmas after the last tag point
                if last_index < len(sentence_lemmas):
                    tagged_lemmas.extend(sentence_lemmas[last_index:])

                # --- Store the relation candidate ---
                relation_candidates.append({
                    'docId': docId,
                    'sentId': sentId,
                    'quantityId': quantityId,
                    'otherId': otherId,
                    'lemmas': tagged_lemmas,
                    'label': is_related
                })

    return pd.DataFrame(relation_candidates)
```

```python
def format_examples(data, word2idx, maxlen):
  X = data.apply(lambda row: [word2idx.get(lemma, word2idx['[UNK]']) for lemma in row['lemmas']], axis=1)
  X = pad_sequences(X, maxlen=maxlen, padding='pre', value=word2idx['[PAD]'])
  y = data['label'].astype(int).values
  return X, y
```

Before returning the model, you should compile it using `'binary_crossentropy'` as the loss function, `'adam'` as the optimizer and `'binary_accuracy'` as a metric to evaluate the model during training

```python
def create_model(vocab_size, maxlen, embedding_dim, filters, kernel_size, hidden_dim):
  model = Sequential()
  model.add(Input(shape=(maxlen,)))
  model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=maxlen))
  model.add(Conv1D(filters=filters, kernel_size=kernel_size, activation='relu'))
  model.add(GlobalMaxPooling1D())
  model.add(Dense(units=hidden_dim, activation='relu'))
  model.add(Dense(units=1, activation='sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
  return model
```
