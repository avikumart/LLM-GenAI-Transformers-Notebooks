```python
# prompt: generate an openai python chat compleation script to interact with custom fine-tuned model using openai models

#!pip install openai

from openai import OpenAI
import openai
from google.colab import userdata
import os

# Replace with your actual API key
os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

def chat_with_custom_model(prompt, model_id):
  """
  Interacts with a custom fine-tuned OpenAI chat model.

  Args:
    prompt: The user's input string.
    model_id: The ID of your fine-tuned model (e.g., "ft:gpt-3.5-turbo-0613:your-org::xxxxxxxx").

  Returns:
    The response from the model.
  """
  try:
    client = OpenAI()
    response = client.chat.completions.create(
        model=model_id,
        messages=[
            {"role": "system", "content": "You are a helpful mental health assistant who guides users to answer their mental health conditions and issues with empathy and kindness."}, # Adjust system message if needed
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
  except Exception as e:
    return f"An error occurred: {e}"
```

```python
# Example usage:
custom_model_id = "ft:gpt-4.1-2025-04-14:personal:omdena-mental-health:BYvcodDC" # Replace with your actual model ID
user_input = "what is anxiety and how to deal with it?"
model_response = chat_with_custom_model(user_input, custom_model_id)
model_response
```
